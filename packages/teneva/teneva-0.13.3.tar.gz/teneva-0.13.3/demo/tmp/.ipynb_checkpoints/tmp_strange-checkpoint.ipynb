```python
import numpy as np
import teneva
np.random.seed(42)
```

```python
d = 5
n = [8] * d
```

```python
func = teneva.FuncDemoRosenbrock(d)
func.set_grid(n, kind='uni')
I_full = teneva.grid_flat(func.n)
Y_full = func.get_f_ind(I_full).reshape(func.n, order='F')
```

```python
def get_real(i):
    return Y_full[tuple(i)]
```

```python
m_trn = int(9.E+0)
I_trn = teneva.sample_lhs(n, m_trn)
Y_trn = np.array([get_real(i) for i in I_trn])
```

```python
m_tst = int(1.E+4)
I_tst = np.vstack([np.random.choice(k, m_tst) for k in n]).T
Y_tst = np.array([get_real(i) for i in I_tst])
```

```python
def poi_set(Y, i, v):
    d = len(Y)
    n = teneva.shape(Y) 
    v_old = teneva.get(Y, i)
    v = v / v_old
    v = abs(v)**(1./d) if abs(v) > 1.E-16 else 1.
    for k in range(d):
        Y[k][0, i[k], 0] = v
```

```python
def poi_add(Y, i, v):
    d = len(Y)
    n = teneva.shape(Y) 
    v_old = teneva.get(Y, i)
    v = v / v_old
    v = abs(v)**(1./d) if abs(v) > 1.E-16 else 1.
    for k in range(d):
        Y[k][0, i[k], 0] = v
```

```python
def poi_max(Y, i, v):
    d = len(Y)
    n = teneva.shape(Y) 
    v_old = teneva.get(Y, i)
    v = v / v_old
    v = abs(v)**(1./d) if abs(v) > 1.E-16 else 1.
    for k in range(d):
        Y[k][0, i[k], 0] = v
```

```python
def _pop_random(x):
    return x.pop(np.random.choice(range(len(x))))

def generate_indices(n):
    d = len(n)
    indices = [list(range(k)) for k in n]
    I = [[_pop_random(indices[i]) for i in range(d)] for _ in range(min(n))]
    return np.array(I, dtype=int)

def build_random(n):
    d = len(n)
    Y = [np.ones([1, k, 1]) for k in n]
    I = generate_indices(n)
    for i in I:
        v = get_real(i)
        s = abs(v) / v if abs(v) > 1.E-16 else v
        v = abs(v)**(1./d) if abs(v) > 1.E-16 else 1.
        for k in range(d):
            Y[k][0, i[k], 0] = v * (1 if k < d-1 else s)
    return Y, I

def build_random_many(n, num=1):
    Y_all = []
    I_all = np.ones((0, len(n)), dtype=int)
    for _ in range(num):
        Y, I = build_random(n)
        Y_all.append(Y)
        I_all = np.vstack((I_all, I))
    return np.array(Y_all), I_all
```

```python
d = 5
q = 3
n = [2**q] * d
n_qtt = [2]*(q*d)
```

```python
func = teneva.FuncDemoRosenbrock(d)
func.set_grid(n, kind='uni')
I_full = teneva.grid_flat(func.n)
Y_full_base = func.get_f_ind(I_full).reshape(func.n, order='F')
Y_full = Y_full_base.reshape([2]*(q*d), order='F')
```

```python
def get_real_tt(i):
    return Y_full_base[tuple(i)]
```

```python
def get_real(i):
    return Y_full[tuple(i)]
```

```python
def qtt_to_tt(i, q):
    d = int(len(i) / q)
    n_qtt = [2]*q
    i_tt = np.zeros(d, dtype=int)
    for k in range(d):
        i_tt[k] = np.ravel_multi_index(i[q*k:q*(k+1)], n_qtt, order='F')
    return i_tt
```

```python
i = [0] * (q*d); i[2] = 1; i[7] = 1
i_tt = qtt_to_tt(i, q)
print(i_tt)
print(get_real_tt(i_tt))
print(get_real(i))
```
