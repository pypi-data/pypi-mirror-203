```python
import numpy as np
import teneva
from time import perf_counter as tpc
np.random.seed(42)
```

```python
def solve(f, a, b, n, m, r, I_tst, Y_tst, y_opt_real):
    opt = {'i_opt': None, 'x_opt': None, 'y_opt': None}
    
    # Приближаемая крестом модифицированная функция:
    def black_box(I):
        # Это список реальных значений функции:
        y = f(I)
        
        # Находим оптимум (индекс i, точка x, значение y):
        s = np.argmin(y)
        if opt['y_opt'] is None or y[s] < opt['y_opt']:
            opt['i_opt'] = I[s, :].copy()
            opt['x_opt'] = teneva.ind_to_poi(opt['i_opt'], a, b, n, 'uni')
            opt['y_opt'] = y[s]
            
        # Преобразуем в экспоненту и возвращаем кресту:
        z = np.exp(-y)
        return z

    # Запуск непосредственно креста:
    t = tpc()
    cache = {}
    Y = teneva.tensor_rand(n, r) # Случайное начальное приближение
    Y = teneva.cross(black_box, Y, m, dr_min=0, dr_max=0, cache=cache, log=False) # dr_max=0 -> не ранг адаптивный крест!
    Y = teneva.truncate(Y, 1.E-12)
    t = tpc() - t # Время расчета
    
    # Формируем использованный крестом набор данных для обучения:
    I_trn, Y_trn = teneva.cache_to_data(cache)

    # Применяем экспоненту к test данным:
    Y_tst = np.exp(-Y_tst)
    
    # Вычисляем ошибку аппроксимации тензора:
    e_trn = teneva.accuracy_on_data(Y, I_trn, Y_trn)
    e_tst = teneva.accuracy_on_data(Y, I_tst, Y_tst)
    
    # Вычисляем ошибку найденного оптимума:
    e_opt = abs(opt['y_opt'] - y_opt_real)
    
    text = func.name + ' ' * (12 - len(func.name)) + '| '
    text += f'e_trn: {e_trn:-7.1e} | '
    text += f'e_tst: {e_tst:-7.1e} | '
    text += f'e_opt: {e_opt:-7.1e} | '
    text += f'time: {t:-8.3f} | '
    text += f'rank: {teneva.erank(Y):-5.1f} | '
    print(text)
    
    return Y, opt
```

```python
d = 5         # Dimension
n = 2**8      # Grid size
r = 3         # Rank
m_trn = 1.E+5 # Train size
m_tst = 1.E+4 # Test size
```

```python
for func in teneva.func_demo_all(d, only_with_min=True):
    if func.name in ['Dixon', 'Qing', 'Rosenbrock', 'Schwefel']: continue # It works with error in this case ((
    
    func.set_grid(n, kind='uni')
    func.build_tst_ind(m_tst)
    
    # Для наглядности мы далее не будем использовать экземпляр класса "func",
    # а просто вытащим из него нужные для расчета и проверки величины:
    f = func.get_f_ind        # Целевая функция (можно передавать батчи индексов)
    a = func.a                # Минимальные значения x
    b = func.b                # Максимальные значения x
    n = func.n                # Размеры мод
    I_tst = func.I_tst_ind    # Тестовые индексы
    Y_tst = func.Y_tst_ind    # Тестовые значения
    y_opt_real = func.y_min   # Точное значение минимума
    
    Y, opt = solve(f, a, b, n, m_trn, r, I_tst, Y_tst, y_opt_real)
```

```python
# Show 1D plots y vs e(-y):

import matplotlib.pyplot as plt

for func in teneva.func_demo_all(d=1, only_with_min=True):
    if func.name in ['Dixon', 'Qing', 'Rosenbrock']: continue
        
    x = np.linspace(func.a[0], func.b[0], 5000)
    y = func.get_f_poi(x.reshape(-1, 1))
    z = np.exp(-y)

    plt.plot(x, y, color='blue')
    plt.plot(x, z, '--', color='orange')
    plt.scatter(func.x_min[0], func.y_min, s=50, c='blue')
    plt.scatter(func.x_min[0], np.exp(-func.y_min), s=50, c='orange')
    plt.title(func.name + ' function (1D)')
    plt.show()
```
